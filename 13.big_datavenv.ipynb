```python
import polars as pl
import pandas as pd
import datetime as dt

import os
import gc
```

```python
ENDERECO_DADOS = r'C:/Users/martins.ryan/an-lise/Base/'

try:
    print('Obtendo arquivos')
    hora_inicio = dt.datetime.now()

    lista_arquivos = os.listdir(ENDERECO_DADOS)

    for arquivo in lista_arquivos:
        print(f'Lendo arquivo: {arquivo}')

        ## df = pd.read_csv(ENDERECO_DADOS + arquivo, sep=';', encoding='iso-8859-1')

        df = pl.read_csv(ENDERECO_DADOS + arquivo, separator=';', encoding='iso-8859-1')

        if 'df_bf' in locals():
            df_bf = pl.concat([df_bf, df])
        else:
            df_bf = df

        del df
    
    df_bf = df_bf.with_columns(
        pl.col('VALOR PARCELA').str.replace(',', '.').cast(pl.Float64)
    )
    df_bf.write_parquet('C:/Users/martins.ryan/an-lise/dados_bronze/df_base.parquet')

    gc.collect()
    
    hora_fim = dt.datetime.now()
    print(f'Tempo todo: {hora_fim - hora_inicio}')

except Exception as e:
    print(f'Erro ao obter dados {e}')
```
