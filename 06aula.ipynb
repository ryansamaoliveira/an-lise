```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
try:
    df = pd.read_csv('03.BaseDPEvolucaoMensalCisp.csv', encoding='iso-8859-1', sep=';')
    # print(df.head())
    print(df.columns)
except Exception as e:
    print(f'ERRO: {e}')
    exit()
```

```python
df_roubo_veiculo = df[['munic', 'roubo_veiculo']]
# print(df_roubo_veiculo.head())

# agrupar por município 
df_roubo_veiculo = df_roubo_veiculo.groupby('munic').sum().reset_index().sort_values(by='roubo_veiculo', ascending=False)
df_roubo_veiculo
```

```python
df_roubo_veiculo_array = np.array(df_roubo_veiculo['roubo_veiculo'])

amplitude = np.max(df_roubo_veiculo_array) - np.min(df_roubo_veiculo_array)

media_roubos = np.mean(df_roubo_veiculo_array)
mediana_roubos = np.median(df_roubo_veiculo_array)

q1 = np.quantile(df_roubo_veiculo_array, 0.25)
q2 = np.quantile(df_roubo_veiculo_array, 0.50, method='weibull')
q3 = np.quantile(df_roubo_veiculo_array, 0.75)

print(f'Q2: {q2} ------ Mediana: {mediana_roubos}')
```
